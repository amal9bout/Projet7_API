```python
from flask import Flask, request, jsonify
import pandas as pd
import joblib

API = Flask(__name__)

# Définir une route API pour prédire le défaut du client
@API.route('/predict', methods=['POST'])
def predict():
    try:
        # Récupérer SK_ID_CURR à partir des données POST
        sk_id_curr = int(request.json['SK_ID_CURR'])

        # Filtrer les données pour obtenir les informations du client
        client_data = df[df['SK_ID_CURR'] == sk_id_curr]

        if client_data.empty:
            return jsonify({'error': 'SK_ID_CURR not found'}), 404

        # Sélectionner les features pour la prédiction (exclure SK_ID_CURR et TARGET)
        features = client_data.drop(['SK_ID_CURR', 'TARGET'], axis=1)

        # Faire la prédiction de probabilité de défaut du client
        pred_proba = model.predict_proba(features)[:, 1]

        # Déterminer la classe en fonction du seuil métier (par exemple, 0.5)
        threshold = 0.5  # Seuil à optimiser pour votre problème métier
        prediction_class = 'Refusé' if pred_proba >= threshold else 'Accepté'

        # Retourner la réponse au format JSON
        return jsonify({
            'SK_ID_CURR': sk_id_curr,
            'Défaut du client (probabilité)': float(pred_proba),
            'Classe': prediction_class
        }), 200

    except Exception as e:
        return jsonify({'error': str(e)}), 500
```
